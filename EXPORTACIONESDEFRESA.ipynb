## Promedio y Desviacion Estandar
- Promedio (media): representa el numero promedio del valor,toneladas y volumen en las diferentes ciudades de colombia 

- Desviación estándar: Esto indica que los valores exportados varian bastante entre los diferentes registros; es decir, algunos periodos o regiones exportaron mucho mas y otros mucho menos.

```python
import pandas as pd

# se carga el archivo Excel
archivo = "Exportacionesagrícolasnotradicionalesytradicionales (1).xlsx"
df = pd.read_excel(archivo, sheet_name="DATA")

# Tomar los primeros 60 datos del archivo
df_60 = df.head(60)

# Mostrar una vista previa
print("Primeros  registros:")
print(df_60.head())

# se calcula los promedios y desviaciones estandar por tipo de exportacion
estadisticas_tradicion = df_60.groupby("Tradición producto")[
    ["Exportaciones en valor (USD FOB)", "Exportaciones en volumen (toneladas)"]
].agg(["mean", "std"]).round(2)

# se calcula los promedios por producto tomados del excel 
promedio_producto = df_60.groupby("Producto")[
    ["Exportaciones en valor (USD FOB)", "Exportaciones en volumen (toneladas)"]
].mean().round(2)

# se calcula el promedios por año
promedio_ano = df_60.groupby("Año")[
    ["Exportaciones en valor (USD FOB)", "Exportaciones en volumen (toneladas)"]
].mean().round(2)

# se calcula los promedios por departamento
promedio_departamento = df_60.groupby("Departamento")[
    ["Exportaciones en valor (USD FOB)", "Exportaciones en volumen (toneladas)"]
].mean().round(2)

# Mostrar resultados
print(" Promedio y desviación estándar por tipo de exportación ")
print(estadisticas_tradicion)

print("Promedio por producto ")
print(promedio_producto)

print("Promedio por año")
print(promedio_ano)

print("Promedio por departamento ")
print(promedio_departamento)
```


## Filtrado avanzado de los accidentes viales

#### En esta sección se implementa un menú interactivo de filtrado que permite a un usuario seleccionar diferentes criterios para analizar los datos segun sean de su interes.

#### El sistema realiza el filtrado a partir de las siguientes opciones:

1. **Por tipo de exportacion: filtra solamente los tipos de exportacion existentes en el archivo.**

2. **Por Volumen: se filtraron las exportaciones con un volumen superior a una tonelada, para excluir valores minimos.**

3. **Por su ubicacion geografica: se analizaron especificamente los departamentos de Cundinamarca, Antioquia, Boyaca y Nariño, por su relevancia en la produccion y exportacion del fruto.**


-El resultado del filtrado avanzado se presenta en formato tabular y se guarda en el archivo “Filtro.xlsx” para su análisis posterior. Este proceso permite comparar y explorar la distribución geográfica y el volumen de exportación de fresas, identificando patrones según el departamento, la magnitud exportada y el tipo de producto (tradicional o no tradicional).



```python
#se importa la libreria 

import pandas as pd

# Cargar el archivo Excel
archivo = "Exportaciones-agrícolas-no-tradicionales-y-tradicionales (1).xlsx"
df = pd.read_excel(archivo, sheet_name="DATA")

# Limpiar nombres de columnas
df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace("  ", " ", regex=False)

# Mostrar menu de opciones
print("FILTRADO AVANZADO DE EXPORTACIONES")
print("1. Departamentos con mayor exportacion")
print("2. Exportaciones No Tradicionales")
print("3. Ingrese un numero de toneladas")
print("4. Salir")

opcion = input("\nIngrese el número de la opción que desea: ")

# Copia base de los primeros 60 datos
df_60 = df.head(60).copy()
datos_filtrados = df_60.copy()

# Filtros
if opcion == "1":
    promedio_toneladas = df_60["Exportaciones en volumen (toneladas)"].mean()
    datos_filtrados = df_60[df_60["Exportaciones en volumen (toneladas)"] > promedio_toneladas]
    print("Filtro aplicado: departamentos con mayor exportaciones")

elif opcion == "2":
    datos_filtrados = df_60[df_60["Tradición producto"].str.contains("No Tradicional", case=False, na=False)]
    print("Filtro aplicado: exportaciones No Tradicionales")


elif opcion == "3":
    try:
        min_ton = float(input("Ingrese el valor minimo de toneladas: "))
        max_ton = float(input("Ingrese el valor maximo de toneladas: "))
        datos_filtrados = df_60[
            (df_60["Ingrese un valor min de toneladas"] >= min_ton) &
            (df_60["Ingrese un valor max de tonelas "] <= max_ton)
        ]
        print(f" Filtro aplicado: entre {min_ton} y {max_ton} toneladas")
    except ValueError:
        print("Entrada invalida. Ingrese numeros validos.")

elif opcion == "4":
    print("Saliendo del programa...")
else:
    print(" Opción no valida.")

# Mostrar resultados y guardar archivo
if opcion in ["1", "2", "3", "4"]:
    print(" RESULTADO FINAL DEL FILTRADO")
    print("...")
    print(datos_filtrados[["Departamento", "Producto", "Tradición producto", "Exportaciones en volumen (toneladas)"]])
    print("...")
    print(f"Total de registros filtrados: {len(datos_filtrados)}")

    archivo_salida = "Filtro_Combinado.xlsx"
    datos_filtrados.to_excel(archivo_salida, index=False)
    print(f" Archivo '{archivo_salida}' guardado correctamente.")
```
